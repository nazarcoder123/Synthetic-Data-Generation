```python
# Import necessary libraries
import pandas as pd
import numpy as np
from ctgan import CTGAN  # Make sure you have 'ctgan' installed: pip install ctgan

# -----------------------------
# Step 1: Create sample data
# -----------------------------
np.random.seed(42)  # For reproducibility

# Create a DataFrame with both numerical and categorical features
data = pd.DataFrame({
    'age': np.random.randint(18, 80, 1000),  # Random integer ages between 18 and 80
    'income': np.random.normal(50000, 15000, 1000),  # Normally distributed income
    'education': np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], 1000),  # Random education levels
    'city': np.random.choice(['New York', 'LA', 'Chicago', 'Houston'], 1000),  # Random cities
    'married': np.random.choice([True, False], 1000)  # Random marital status
})

# Ensure no negative income values (absolute value of income)
data['income'] = np.abs(data['income'])

# Display shape and a sample of the original data
print("Original Data Shape:", data.shape)
print("\nOriginal Data Sample:")
print(data.head())

# -----------------------------
# Step 2: Initialize CTGAN
# -----------------------------
ctgan = CTGAN(epochs=100, batch_size=500, verbose=True)  # Set training parameters

# -----------------------------
# Step 3: Specify discrete (categorical) columns
# -----------------------------
discrete_columns = ['education', 'city', 'married']

# -----------------------------
# Step 4: Train the CTGAN model
# -----------------------------
print("\nTraining CTGAN...")
ctgan.fit(data, discrete_columns)

# -----------------------------
# Step 5: Generate synthetic data
# -----------------------------
print("\nGenerating synthetic data...")
synthetic_data = ctgan.sample(1000)

# Display synthetic data shape and sample
print("\nSynthetic Data Shape:", synthetic_data.shape)
print("\nSynthetic Data Sample:")
print(synthetic_data.head())

# -----------------------------
# Step 6: Compare distributions
# -----------------------------
print("\n=== COMPARISON ===")

# Compare education distribution between original and synthetic datasets
print("\nOriginal Education Distribution:")
print(data['education'].value_counts(normalize=True))
print("\nSynthetic Education Distribution:")
print(synthetic_data['education'].value_counts(normalize=True))

# Compare descriptive statistics of age
print("\nOriginal Age Statistics:")
print(data['age'].describe())
print("\nSynthetic Age Statistics:")
print(synthetic_data['age'].describe())
```

```python
from mimesis import Generic  # Import the Generic class from the mimesis library to generate synthetic data
from mimesis.locales import Locale  # Import Locale enum to specify locale for fake data
import pandas as pd  # Import pandas for data manipulation and creating DataFrame

# Set locale (British English)
fake = Generic(locale=Locale.EN_GB)  # Initialize the Generic faker object with British English locale

# Generate synthetic data
df = pd.DataFrame([  # Create a pandas DataFrame from a list of dictionaries
    {
        "name": fake.person.full_name(),  # Generate a full name using the fake person generator
        "country": fake.address.country(),  # Generate a country name using the fake address generator
        "birthday": fake.datetime.date(),  # Generate a random date as birthday using the fake datetime generator
        "email": fake.person.email(),  # Generate an email address using the fake person generator
        "address": fake.address.address(),  # Generate a full address using the fake address generator
    }
    for _ in range(100)  # Repeat the above dictionary generation 100 times to create 100 rows
])
```
